```python
# Adjust dates so we can see how fast an outbreak is spreading relative to other countries.
# Adjust the dataset so all dates start from a common 0 day so it's possible to compare the slopes
# between countries
ebola = pd.read_csv('../data/country_timeseries.csv', index_col='Date', parse_dates=['Date'])
ebola.head().iloc[:, :4]
```

```python
ebola.tail().iloc[:, :4]
```

```python
new_idx = pd.date_range(ebola.index.min(), ebola.index.max())
new_idx
```

```python
new_idx = reversed(new_idx)
```

```python
ebola = ebola.reindex(new_idx)
```

```python
display(ebola.head().iloc[:, :4])
display(ebola.tail().iloc[:, :4])
```

```python
# calculate the difference between earliest date in our data set
# and the earliest valid (non-missing) date in each column.
last_valid = ebola.apply(pd.Series.last_valid_index)
last_valid
```

```python
# earliest date in dataset
earliest_date = ebola.index.min()
earliest_date
```

```python
# subtract this date from each of our last_valid dates
shift_values = last_valid - earliest_date
shift_values
```

```python
# iterate through each column and use the shift method to shift columns down
# by the corresponding value in shift_values
ebola_dict = {}
for idx, col in enumerate(ebola):
    d = shift_values[idx].days
    shifted = ebola[col].shift(d)
    ebola_dict[col] = shifted
ebola_shift = pd.DataFrame(ebola_dict)
ebola_shift = ebola_shift[ebola.columns]
ebola_shift.tail()
```

```python
# remove the indices and assign the correct index, which is the Day
# day represents the first day of an outbreak of the given country
ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Day'], axis=1)
ebola_shift.tail()
```

```python
fig, ax = plt.subplots()
ax = ebola_shift.plot(ax=ax)
ax.legend(fontsize=7, loc=2, borderaxespad=0.)
plt.show()
```
